```python
import numpy as np
import pandas as pd
```

```python
data = {'Name': ['Amit', 'Riya', 'Karan', 'Neha', 'Raj'],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Age': [25, 30, 35, 28, 22],
        'City': ['Delhi', 'Mumbai', 'Kolkata', 'Chennai', 'Delhi'],
        'Education_Level': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor']}

df = pd.DataFrame(data)
df
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender_le'] = le.fit_transform(df['Gender'])
df['City_le'] = le.fit_transform(df['City'])
df[['Gender', 'Gender_le', 'City', 'City_le']]
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)
city_encoded = ohe.fit_transform(df[['City']])
name = ohe.get_feature_names_out(['City'])
city_encoded_df = pd.DataFrame(city_encoded, columns=name)
city_encoded_df
```

```python
df_final = pd.get_dummies(df, columns=['City'])
df_final
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

edu_order = [['High School', 'Bachelor', 'Master', 'PhD']]
oe = OrdinalEncoder(categories=edu_order)
df['Education_Encoded'] = oe.fit_transform(df[['Education_Level']])
df[['Education_Level', 'Education_Encoded']]
```
