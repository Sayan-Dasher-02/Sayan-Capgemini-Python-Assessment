```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, LeaveOneOut, StratifiedKFold, cross_val_score
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df['target'] = data.target
df
```

```python
x = df.drop('target', axis=1)
y = df['target']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model_loo = RandomForestClassifier()
```

```python
loo = LeaveOneOut()
scores_loo = cross_val_score(model_loo, xtrain, ytrain, cv=loo)
print('LOOCV TRaining Efficiency =', scores_loo.mean())
```

```python
model_loo.fit(xtrain, ytrain)
```

```python
print('LOOCV Testing Efficiency =', model_loo.score(xtest, ytest))
```

```python
#Stratified Cross Validation
model_strat = RandomForestClassifier()
skf = StratifiedKFold(n_splits=5)
scores_strat = cross_val_score(model_strat, xtrain, ytrain, cv=skf)
print('Stratified CV Training Efficiency =', scores_strat.mean())
```

```python
model_strat.fit(xtrain, ytrain)
```

```python
print('Stratified CV Testing Efficiency =', model_strat.score(xtest, ytest))
```
