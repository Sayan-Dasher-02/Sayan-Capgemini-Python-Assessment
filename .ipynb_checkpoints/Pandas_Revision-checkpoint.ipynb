```python
def convert_price(x):
  if isinstance(x, str) and 'M' in x:
    return float(x.replace('M', '')) * 1_000_000
  else:
    return float(x)
```

```python
df['Price']=df['Price'].apply(convert_price)
df
```

```python
df['Price'].isnull().sum()/len(df['Price'])*100
```

```python
df['Price']=df['Price'].fillna(df['Price'].mean())
df
```

Data Preprocessing

```python
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Puri', 'Bhubaneswar', 'Cuttack', 'Khordha']])
df['City'] = oe.fit_transform(df[['City']])
df
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Garage'] = le.fit_transform(df['Garage'])
df['Furnished'] = le.fit_transform(df['Furnished'])
df
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)
property_encoded = ohe.fit_transform(df[['Property_Type']])
name = ohe.get_feature_names_out(['Property_Type'])
property_encoded_df = pd.DataFrame(property_encoded, columns=name)
property_encoded_df
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)
loan_encoded = ohe.fit_transform(df[['Loan_Status']])
name = ohe.get_feature_names_out(['Loan_Status'])
loan_encoded_df = pd.DataFrame(loan_encoded, columns=name)
loan_encoded_df
```

```python
df = pd.get_dummies(df, columns=['Property_Type', 'Loan_Status'], dtype=int)
df
```

```python
df['Bedrooms']=df['Bedrooms'].astype(int)
df['Bathrooms']=df['Bathrooms'].astype(int)
```

```python
df['Total_rooms'] = df['Bedrooms'] + df['Bathrooms']
df
```

```python
df['Price_per_Sqft'] = df['Price'] / df['Area_sqft']
df
```

```python
df['House_Age'] = 2026 - df['Year_Built']
df
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
```

```python
outliers
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df['Price']))
```

```python
outliers = df[z > 3]
outliers
```

```python
outliers[outliers['Area_sqft']>3000]
```

```python
# df = df[df['Price'] < 100_000_000] ---- This method helps to remove the outliers from the dataset
```

```python
upper_limit = 100_000_000
df['Price'] = np.where(df['Price'] > upper_limit, upper_limit, df['Price'])
#This method changes the outlier to the upeerlimit value or the expected value if the condition satisfies
```

```python
Q1 = df['Year_Built'].quantile(0.25)
Q3 = df['Year_Built'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Year_Built'] < lower_bound) | (df['Year_Built'] > upper_bound)]
```
