```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Students Social Media Addiction.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Academic Level distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Academic_Level', data=df)
plt.title("Academic Level Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
# Country distribution (top 10)
plt.figure(figsize=(10,5))
top_countries = df['Country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Countries in Dataset")
plt.xticks(rotation=30)
plt.show()
```

```python
# Most Used Platform
plt.figure(figsize=(8,5))
sns.countplot(x='Most_Used_Platform', data=df, order=df['Most_Used_Platform'].value_counts().index)
plt.title("Most Used Social Media Platform")
plt.xticks(rotation=30)
plt.show()
```

```python
#Line Chart

plt.figure(figsize=(10,5))
avg_time_age = df.groupby('Age')['Avg_Daily_Usage_Hours'].mean()
plt.plot(avg_time_age.index, avg_time_age.values, marker='o', color='teal')
plt.title("Average Time Spent vs Age")
plt.xlabel("Age")
plt.ylabel("Average Time (hours/day)")
plt.xticks(avg_time_age.index)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Relationship_Status', y='Mental_Health_Score', data=df)
plt.title("Mental Health Score by Relationship Status")
plt.xticks(rotation=30)
plt.show()
```

```python
#Pie Chart

plt.figure(figsize=(6,6))
relation_counts = df['Relationship_Status'].value_counts()
plt.pie(relation_counts, labels=relation_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Relationship Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Sleep_Hours_Per_Night', y='Addicted_Score', hue='Gender', data=df)
plt.title("Sleep Hours vs Addiction Score")
plt.show()
```

```python
#Histogram

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
num = df.select_dtypes(include=['int64','float64'])
corr = num.corr()
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.title("Correlation Heatmap")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()
```
