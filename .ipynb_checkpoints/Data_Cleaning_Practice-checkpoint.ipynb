Phone Number Cleaning

```python
df['Phone_Number'].value_counts().head(20)
```

```python
df['Cleaned_Phone'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
```

```python
#Extract only country code(first 1-3 digits if number>10 digits)
def get_country_code(number):
    if pd.isnull(number):
        return np.nan
    if len(number) <= 10:
        return np.nan
    return number[:-10] #Everything before the last 10 digits

df['Country_Code'] = df['Cleaned_Phone'].apply(get_country_code)
```

```python
df['Country_Code'].isnull().sum()
```

```python
def get_local_number(number):
    if pd.isnull(number) or len(number) < 10:
        return np.nan
    return number[:-10]

df['Local_Number'] = df['Cleaned_Phone'].apply(get_local_number)
```

```python
df['Is_Valid_Number'] = df['Local_Number'].notnull()
```

```python
unique_salaries
```

```python
import numpy as np
```

```python
def salary_format_type(s):
    s = str(s).strip()
    if s == 'nan':
        return 'Missing'
    elif any(c.isalpha() for c in s):
        return 'Contains letters(k, M, etc.)'
    elif any(c in '₹$,' for c in s):
        return 'Currency / Comma'
    elif s.replace('.', '', 1).isdigit():
        return 'Numeric'
    else:
        return 'Other / Invalid'

Salary_Format = salary_series.apply(salary_format_type)

#check counts per format type
Salary_Format.value_counts()
```

```python
currency_comma_mask = salary_str.str.contains(r'[₹$€,]', regex=True)
currency_comma_mask
```

```python
unique_currency_comma = salary_str[currency_comma_mask].unique()
unique_currency_comma
```

```python
print('Total unique currency/comma formats:',len(unique_currency_comma))
print('Sample unique currency/comma formats:')
print(unique_currency_comma[:53])
```
