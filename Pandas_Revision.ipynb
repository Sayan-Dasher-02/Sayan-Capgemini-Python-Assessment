```python
def convert_price(x):
  if isinstance(x, str) and 'M' in x:
    return float(x.replace('M', '')) * 1_000_000
  else:
    return float(x)
```

```python
df['Price']=df['Price'].apply(convert_price)
df
```

```python
df['Price'].isnull().sum()/len(df['Price'])*100
```

```python
df['Price']=df['Price'].fillna(df['Price'].mean())
df
```
