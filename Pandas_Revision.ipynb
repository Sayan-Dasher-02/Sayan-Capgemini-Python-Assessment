```python
def convert_price(x):
  if isinstance(x, str) and 'M' in x:
    return float(x.replace('M', '')) * 1_000_000
  else:
    return float(x)
```

```python
df['Price']=df['Price'].apply(convert_price)
df
```

Data Preprocessing

```python
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Puri', 'Bhubaneswar', 'Cuttack', 'Khordha']])
df['City'] = oe.fit_transform(df[['City']])
df
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Garage_le'] = le.fit_transform(df['Garage'])
df['Furnished_le'] = le.fit_transform(df['Furnished'])
df[['Garage', 'Garage_le', 'Furnished', 'Furnished_le']]
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)
property_encoded = ohe.fit_transform(df[['Property_Type']])
name = ohe.get_feature_names_out(['Property_Type'])
property_encoded_df = pd.DataFrame(property_encoded, columns=name)
property_encoded_df
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)
loan_encoded = ohe.fit_transform(df[['Loan_Status']])
name = ohe.get_feature_names_out(['Loan_Status'])
loan_encoded_df = pd.DataFrame(loan_encoded, columns=name)
loan_encoded_df
```

```python
df_final = pd.get_dummies(df, columns=['Property_Type', 'Loan_Status'])
df_final
```
