```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentPerformanceFactors.csv')
df
```

```python
df['School_Type'].unique()
```

```python
private=df.groupby('School_Type').get_group('Private')
private
public=df.groupby('School_Type').get_group('Public')
public
```

```python
sample_size=100
private_sample=private['Attendance'].sample(sample_size)
private_sample
public_sample=public['Attendance'].sample(sample_size)
public_sample
```

```python
pop_mean_private=private['Attendance'].mean()
pop_mean_private
pop_mean_public=public['Attendance'].mean()
pop_mean_public
```

```python
sample_mean_private=private_sample.mean()
sample_mean_private
sample_mean_public=public_sample.mean()
sample_mean_public
```

```python
error_private=pop_mean_private-sample_mean_private
error_private
error_public=pop_mean_public-sample_mean_public
error_public
```

```python
#For Private Schools
plt.title('Scatter Plot for Private School')
plt.scatter(range(1,sample_size+1),private_sample)
plt.axhline(y=pop_mean_private,color='red',label='population mean private')
plt.axhline(y=sample_mean_private,color='green',label='private school')
plt.legend()
plt.show()
```

```python
#For Public Schools
plt.title('Scatter Plot for Public School')
plt.scatter(range(1,sample_size+1),public_sample)
plt.axhline(y=pop_mean_public,color='red',label='population mean public')
plt.axhline(y=sample_mean_public,color='green',label='public school')
plt.legend()
plt.show()
```
