```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Students Social Media Addiction.csv')
df.head()
```

```python
#Line Chart

plt.figure(figsize=(10,5))
avg_time_age = df.groupby('Age')['Avg_Daily_Usage_Hours'].mean()
plt.plot(avg_time_age.index, avg_time_age.values, marker='o', color='teal', linewidth=1)
plt.title("Average Time Spent vs Age")
plt.xlabel("Age")
plt.ylabel("Average Time (hours/day)")
plt.xticks(avg_time_age.index)
plt.show()
```

```python
#Bar Chart

plt.figure(figsize=(10,5))
platform_counts = df['Most_Used_Platform'].value_counts()
plt.bar(platform_counts.index, platform_counts.values, color='mediumseagreen', edgecolor='black')
plt.title("Most Preferred Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.xticks(rotation=30)
plt.show()
```

```python
#Pie Chart

plt.figure(figsize=(6,6))
relation_counts = df['Relationship_Status'].value_counts()
plt.pie(relation_counts, labels=relation_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Relationship Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Avg_Daily_Usage_Hours'], df['Sleep_Hours_Per_Night'], color='orange', edgecolor='black', alpha=0.7)
plt.title("Daily Usage vs Sleep Hours")
plt.xlabel("Avg Daily Usage (hours)")
plt.ylabel("Sleep Hours per Night")

plt.show()
```

```python
#Histogram

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```
