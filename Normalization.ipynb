```python
import numpy as np
import pandas as pd
```

```python
data = {'Name': ['S1', 'S2', 'S3', 'S4'],
        'Height': [162, 171, 155, 165],
        'Weight': [63, 60, 50, 62],
        'Speed': [5, 3, 7, 2]}

df=pd.DataFrame(data)
df
```

```python
#L2 Normalization

from sklearn.preprocessing import Normalizer

norm = Normalizer()
num_data = df[['Height', 'Weight', 'Speed']]
df_norm = norm.fit_transform(num_data)
df = pd.DataFrame(df_norm, columns = num_data.columns)
df
```

```python
#Standard Scaler Normalization

from sklearn.preprocessing import StandardScaler

norm = StandardScaler()
num_data = df[['Height', 'Weight', 'Speed']]
df_norm = norm.fit_transform(num_data)
df = pd.DataFrame(df_norm, columns = num_data.columns)
df
```
